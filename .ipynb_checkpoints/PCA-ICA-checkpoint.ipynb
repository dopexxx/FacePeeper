```python
# Performs and displays PCA and ICA on Different set of stimuli

class components:
    
    def __init__(self,direc, numComp,num):
    
        # Imports
        import numpy as np
        from PIL import Image
        import matplotlib.pyplot as plt
        from sklearn.decomposition import FastICA, PCA
        from skimage.color import rgb2gray
        import scipy.misc as smp
        import glob
    
        self.imSize = 128 # Default
        self.numComp = numComp
        self.txt = 'Condition ' + num
        
        # Bad solution to avoid too many open files
        tmp = []
        for filename in glob.glob(direc+'/Images/*.png'):
            im = Image.open(filename)
            tmp.append(im)
            im.close()

        image_list = []
        self.data = np.zeros([len(tmp),self.imSize*self.imSize])
        for ind,filename in enumerate(glob.glob(direc+'/Images/*.png')): 
            im= Image.open(filename)
            image_list.append(im)
            self.data[ind,:] = np.reshape(rgb2gray(np.array(im)),(self.imSize*self.imSize)) # Including greyscale conversion
            im.close()
            


    
    def pca(self): # Compute PCA
        self.p = PCA(n_components = self.numComp)
        a = self.p.fit(self.data) # Necessary??
        self.scores = self.p.components_           



    def ica(self): # Compute ICA
        self.i = FastICA(n_components = self.numComp)
        a = self.i.fit(self.data)  # Reconstruct signals
        self.scores = self.i.components_  # Get estimated mixing matrix
        
        
    def plotting(self):
        fig = plt.figure()
        eigenDirections = np.zeros([self.imSize,self.imSize,self.numComp])
        for pc in range(0,self.numComp):
            eigenDirections[:,:,pc] = np.reshape(self.scores[pc,:],(self.imSize,self.imSize))
            a = fig.add_subplot(1,self.numComp,pc+1)
            b = smp.toimage(eigenDirections[:,:,pc])
            plt.imshow(b)
            plt.axis('off')
        plt.suptitle(self.txt)
        plt.savefig('/Network/Servers/mac0.cns.ox.ac.uk/Volumes/Data/Users/born/Documents/VisBack-master/Scripts/Results/ICA/'+self.txt)
        plt.show()
```

```python
s = '/Network/Servers/mac0.cns.ox.ac.uk/Volumes/Data/Users/born/Documents/VisBack-master/Stimuli/'
a = '30Locs/'
num = 'b)-4_Comp'
c = components(s+a,4,num)
c.pca()
c.plotting()
c.ica()
c.plotting()
```
